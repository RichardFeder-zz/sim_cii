```python
iNeff = 0
i = 2
sigI = sigI_arr[i]
j = 6
sig_th = sig_th_arr[j]

Itrue_all0 = np.load(datadir + 'Itrue_all%d_Lb20.npy'%iNeff)
Itrue_targ0 = np.load(datadir + 'Itrue_targ%d_Lb20.npy'%iNeff)
Itrue_all1 = np.load(datadir + 'Itrue_all%d_Lb50.npy'%iNeff)
Itrue_targ1 = np.load(datadir + 'Itrue_targ%d_Lb50.npy'%iNeff)
Itrue_all2 = np.load(datadir + 'Itrue_all%d_Lb100.npy'%iNeff)
Itrue_targ2 = np.load(datadir + 'Itrue_targ%d_Lb100.npy'%iNeff)

Iobs_all = Itrue_all + np.random.normal(0, sigI, Itrue_all.shape)
print('sigI = %.0e, sig_th = %.1f'%(sigI, sig_th))

fig, ax = plt.subplots(2,3,figsize=(18,8))

pos = [(0,0), (0,1), (0,2), (1,0), (1,1), (1,2)]
count = 0
Iobs_all0 = Itrue_all0 + np.random.normal(0, sigI, Itrue_all.shape)
Iobs_all1 = Itrue_all1 + np.random.normal(0, sigI, Itrue_all.shape)
Iobs_all2 = Itrue_all2 + np.random.normal(0, sigI, Itrue_all.shape)

Ipred_targ_arr = np.load(datadir + 'Ipred_targ_arr%d%d%d.npy'%(iNeff, i, j))
Ipred_targ_arr0 = np.load(datadir + 'Ipred_targ_arr%d%d%d_Lb20.npy'%(iNeff, i, j))
Ipred_targ_arr1 = np.load(datadir + 'Ipred_targ_arr%d%d%d_Lb50.npy'%(iNeff, i, j))
Ipred_targ_arr2 = np.load(datadir + 'Ipred_targ_arr%d%d%d_Lb100.npy'%(iNeff, i, j))

for jtarg in [3,4,5,6]:
    j_idx = line_use.index(jtarg)
    idx_vec, z_min_vec, z_max_vec, name_vec = survey_band_bins(jtarg, 'survey1')
    for _, (idx,z_min,z_max,name) in enumerate(zip(idx_vec, z_min_vec, z_max_vec, name_vec)):
        if count < len(pos):
            ii, jj = pos[count]
        count += 1
        hmin = np.max(Itrue_targ[j_idx, :, idx]) / 100
        hmax1 = np.max(Ipred_targ_arr[:, j_idx, :, idx])
        hmax2 = np.max(Itrue_targ[j_idx, :, idx])
        hmax = max(hmax1, hmax2)
        Ibinedges = np.arange(np.log10(hmin),np.log10(hmax) + 0.2 ,0.2)
        Ibinedges = 10**Ibinedges
        I_all = (Ibinedges[1:] + Ibinedges[:-1]) / 2
        dIbins = Ibinedges[1:] - Ibinedges[:-1]

        h_true, _ = np.histogram(Itrue_targ0[j_idx, :, idx].flatten(), bins = Ibinedges)
        h_true = h_true / len(Itrue_targ0[j_idx, :, idx].flatten()) / dIbins
        ax[ii][jj].plot(I_all, h_true, 'c-', lw=3, alpha=0.5, label = 'true 20%')
        h_true, _ = np.histogram(Itrue_targ1[j_idx, :, idx].flatten(), bins = Ibinedges)
        h_true = h_true / len(Itrue_targ1[j_idx, :, idx].flatten()) / dIbins
        ax[ii][jj].plot(I_all, h_true, 'y-', lw=3,alpha=0.5, label = 'true 50%')
        h_true, _ = np.histogram(Itrue_targ2[j_idx, :, idx].flatten(), bins = Ibinedges)
        h_true = h_true / len(Itrue_targ2[j_idx, :, idx].flatten()) / dIbins
        ax[ii][jj].plot(I_all, h_true, 'm-', lw=3,alpha=0.5, label = 'true 100%')

#         h_obs, _ = np.histogram(Iobs_all0[:, idx].flatten(), bins = Ibinedges)
#         h_obs = h_obs / len(Iobs_all0[:, idx].flatten()) / dIbins
#         ax[ii][jj].plot(I_all, h_obs, '-', lw=3, c = 'grey', label = 'observed 20%')
#         h_obs, _ = np.histogram(Iobs_all1[:, idx].flatten(), bins = Ibinedges)
#         h_obs = h_obs / len(Iobs_all1[:, idx].flatten()) / dIbins
#         ax[ii][jj].plot(I_all, h_obs, '--', lw=3, c = 'grey', label = 'observed 50%')
#         h_obs, _ = np.histogram(Iobs_all2[:, idx].flatten(), bins = Ibinedges)
#         h_obs = h_obs / len(Iobs_all2[:, idx].flatten()) / dIbins
#         ax[ii][jj].plot(I_all, h_obs, ':', lw=3, c = 'grey', label = 'observed 100%')

#         h_pred_arr = np.zeros([N_sim, len(Ibinedges) - 1])
#         for n_idx in range(N_sim):
#             dat_pred = Ipred_targ_arr[n_idx, j_idx, :, idx].flatten()
#             h_pred, _ = np.histogram(dat_pred, bins = Ibinedges)
#             h_pred = h_pred / len(dat_pred) / dIbins
#             h_pred_arr[n_idx, :] = h_pred
#         ax[ii][jj].errorbar(I_all, np.mean(h_pred_arr, axis = 0), \
#                               np.std(h_pred_arr, axis = 0),  c = 'r', \
#                               fmt = '.', capsize = 4, lw = 3, 
#                                 label = 'No line ratio \nerror rec.')
#         ax[ii][jj].plot(I_all, np.mean(h_pred_arr, axis = 0), 'r.', markersize=5)

        h_pred_arr = np.zeros([N_sim, len(Ibinedges) - 1])
        for n_idx in range(N_sim):
            dat_pred = Ipred_targ_arr0[n_idx, j_idx, :, idx].flatten()
            h_pred, _ = np.histogram(dat_pred, bins = Ibinedges)
            h_pred = h_pred / len(dat_pred) / dIbins
            h_pred_arr[n_idx, :] = h_pred
        ax[ii][jj].errorbar(I_all*0.95, np.mean(h_pred_arr, axis = 0), \
                              np.std(h_pred_arr, axis = 0),  c = 'c', \
                              fmt = '.', capsize = 4, lw = 3, label = '20% SLED bias')
        ax[ii][jj].plot(I_all*0.95, np.mean(h_pred_arr, axis = 0), 'c.', markersize=5)

        h_pred_arr = np.zeros([N_sim, len(Ibinedges) - 1])
        for n_idx in range(N_sim):
            dat_pred = Ipred_targ_arr1[n_idx, j_idx, :, idx].flatten()
            h_pred, _ = np.histogram(dat_pred, bins = Ibinedges)
            h_pred = h_pred / len(dat_pred) / dIbins
            h_pred_arr[n_idx, :] = h_pred
        ax[ii][jj].errorbar(I_all*1.05, np.mean(h_pred_arr, axis = 0), \
                              np.std(h_pred_arr, axis = 0),  c = 'y', \
                              fmt = '.', capsize = 4, lw = 3, label = '50% SLED bias')
        ax[ii][jj].plot(I_all*1.05, np.mean(h_pred_arr, axis = 0), 'y.', markersize=5)

        h_pred_arr = np.zeros([N_sim, len(Ibinedges) - 1])
        for n_idx in range(N_sim):
            dat_pred = Ipred_targ_arr2[n_idx, j_idx, :, idx].flatten()
            h_pred, _ = np.histogram(dat_pred, bins = Ibinedges)
            h_pred = h_pred / len(dat_pred) / dIbins
            h_pred_arr[n_idx, :] = h_pred
        ax[ii][jj].errorbar(I_all*1.05, np.mean(h_pred_arr, axis = 0), \
                              np.std(h_pred_arr, axis = 0),  c = 'm', \
                              fmt = '.', capsize = 4, lw = 3, label = '100% SLED bias')
        ax[ii][jj].plot(I_all*1.05, np.mean(h_pred_arr, axis = 0), 'm.', markersize=5)

        _,I_vec_targ = Ivox_from_zsrc([[(z_min + z_max)/2]], dth,
                                      nu_binedges, line_use, [jtarg])
        ax[ii][jj].axvline(I_vec_targ[I_vec_targ!=0], c= 'g', lw=3, ls='--', label = '$I_*$')
        
        ax[ii][jj].axvline(sigI, c= 'b', ls='--', lw=3, label = '$\sigma_n$')
        ax[ii][jj].set_xscale('log')
        ax[ii][jj].set_yscale('log')
        if ii==1:
            ax[ii][jj].set_xlabel('I [Jy / sr]',fontsize = 15)
        if jj==0:
            ax[ii][jj].set_ylabel('P(I) [(Jy / sr)$^{-1}$]', fontsize = 15)
        ax[ii][jj].set_title(r'%s $\langle z \rangle$ = %.2f'\
                                         %(name, (z_min + z_max)/2), fontsize = 20)
        ax[ii][jj].tick_params(axis='both', which='major', labelsize=15)

ax[0][0].legend(loc = 3, fontsize = 14)
plt.tight_layout()
plt.savefig('plots/paperfig/gen_VID_bias.pdf', bbox_inches = 'tight', dpi = 300)
```
